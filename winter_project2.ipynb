```python
import numpy as np

def string_to_vector(s):
    # Remove square brackets if present
    s = s.replace('[', '').replace(']', '')
    # Split by spaces and convert each number to float
    return np.array([float(x) for x in s.split()], dtype=np.float32)

# Example usage with your dataset
X_train_array = np.array([string_to_vector(x) for x in X_train])
print(X_train_array.shape)  # should be (num_samples, vector_dim)
X_test_array=np.array([string_to_vector(x) for x in X_test])




X_train_final = np.vstack(X_train_array).astype(np.float32)
print(X_train_final.dtype)
X_test_final = np.vstack(X_test_array).astype(np.float32)
print(X_test_final.dtype)
```

```python
from transformers import AutoTokenizer
import os

def load_deberta_v3_tokenizer(model_name="microsoft/deberta-v3-base"):
    try:
        # Try the normal fast tokenizer first
        tokenizer = AutoTokenizer.from_pretrained(model_name)
    except (AttributeError, ImportError):
        # Fallback: force slow tokenizer
        tokenizer = AutoTokenizer.from_pretrained(model_name, use_fast=False)
    return tokenizer

# Load tokenizer safely
tokenizer = load_deberta_v3_tokenizer()

# Test
print(tokenizer.tokenize("Hello world!"))

```

```python
%pip install torch

import torch
from torch.utils.data import Dataset

class TextDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, max_len=128):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_len = max_len
        
    def __len__(self):
        return len(self.texts)
    
    def __getitem__(self, idx):
        text = self.texts[idx]
        label = self.labels[idx]
        encoding = self.tokenizer(
            text,
            truncation=True,
            padding='max_length',
            max_length=self.max_len,
            return_tensors='pt'
        )
        return {
            "input_ids": encoding["input_ids"].squeeze(0),
            "attention_mask": encoding["attention_mask"].squeeze(0),
            "labels": torch.tensor(label, dtype=torch.long)
        }

train_dataset = TextDataset(X_train, y_train, tokenizer)
val_dataset   = TextDataset(X_test, y_test, tokenizer)
```
